```python
import os, sys
import pandas as pd
import numpy as np

us_data = pd.read_csv("../output/US_youtube_trending_data/part-00000")
gb_data = pd.read_csv("../output/GB_youtube_trending_data/part-00000")
ca_data = pd.read_csv("../output/CA_youtube_trending_data/part-00000")
jp_data =  pd.read_csv("../output/JP_youtube_trending_data/part-00000")
```

## Analysis 
### The most popular Youtube video  during 2020 ~ 2023 in US

```python
a1 = us_data.groupby('name').agg({'rank_count': 'sum'}).reset_index()
a1_gb = gb_data.groupby('name').agg({'rank_count': 'sum'}).reset_index()
a1_ca = ca_data.groupby('name').agg({'rank_count': 'sum'}).reset_index()
a1_jp = jp_data.groupby('name').agg({'rank_count': 'sum'}).reset_index()
a1.head(10)
```

```python
import matplotlib.pyplot as plt
from sklearn import preprocessing
plt.figure().set_figwidth(25)
plt.plot(a1['name'], a1['rank_count'], color='blue', label='US')
plt.plot(a1_gb['name'], a1_gb['rank_count'], color='red', label='UK')
plt.plot(a1_ca['name'], a1_ca['rank_count'], color='green', label='CA')
plt.plot(a1_jp['name'], a1_jp['rank_count'], color='black', label='JP')

plt.xticks(rotation=75)
plt.title("The most popular Youtube video  during 2020 ~ 2023") 
plt.xlabel("The video category")
plt.ylabel("Count of daily top 10 popular video (times)")
#plt.gca().legend(('Rent price'))
plt.legend()
plt.show()
```
